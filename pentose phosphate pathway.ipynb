```python
import cobra
```

```python
from cobra import Model,Reaction,Metabolite
```

```python
# Glycolysis pathway

model=Model('example')

Glucose_6_phosphate_dehydrogenase=Reaction('Glucose_6_phosphate_dehydrogenase')
Glucose_6_phosphate_dehydrogenase.name='Glucose_6_phosphate_dehydrogenase'
Glucose_6_phosphate_dehydrogenase.lower_bound=0
Glucose_6_phosphate_dehydrogenase.upper_bound=1000


Gluconolactonase=Reaction('Gluconolactonase')
Gluconolactonase.name='Gluconolactonase'
Gluconolactonase.lower_bound=-1000
Gluconolactonase.upper_bound=1000


p6_phosphogluconate_dehydratenase=Reaction('p6_phosphogluconate_dehydratenase')
p6_phosphogluconate_dehydratenase.name='p6_phosphogluconate_dehydratenase'
p6_phosphogluconate_dehydratenase.lower_bound=0
p6_phosphogluconate_dehydratenase.upper_bound=1000


ribulose_5_phosphate_3_Epimerase=Reaction('ribulose_5_phosphate_3_Epimerase')
ribulose_5_phosphate_3_Epimerase.name='ribulose_5_phosphate_3_Epimerase'
ribulose_5_phosphate_3_Epimerase.lower_bound=0
ribulose_5_phosphate_3_Epimerase.upper_bound=1000


ribulose_5_phosphate_isomerase=Reaction('ribulose_5_phosphate_isomerase')
ribulose_5_phosphate_isomerase.name='ribulose_5_phosphate_isomerase'
ribulose_5_phosphate_isomerase.lower_bound=0
ribulose_5_phosphate_isomerase.upper_bound=1000



transaldolase=Reaction('transaldolase')
transaldolase.name='transaldolase'
transaldolase.lower_bound=0
transaldolase.upper_bound=1000


# xylulose_5_phospate=Reaction('xylulose_5_phospate')
# xylulose_5_phospate.name='xylulose_5_phospate'
# xylulose_5_phospate.lower_bound=0
# xylulose_5_phospate.upper_bound=1000


transletolase1=Reaction('transletolase1')
transletolase1.name='transletolase'
transletolase1.lower_bound=0
transletolase1.upper_bound=1000

transletolase2=Reaction('transletolase2')
transletolase2.name='transletolase'
transletolase2.lower_bound=0
transletolase2.upper_bound=1000


transaldolase3=Reaction('transaldolase3')
transaldolase3.name='transaldolase4'
transaldolase3.lower_bound=0
transaldolase3.upper_bound=1000

transaldolase4=Reaction('transaldolase4')
transaldolase4.name='transaldolase4'
transaldolase4.lower_bound=0
transaldolase4.upper_bound=1000


Glu1=Reaction('Glu1')
Glu1.name='Glu1'
Glu1.lower_bound=-1000
Glu1.upper_bound=1000

Glu2=Reaction('Glu2')
Glu2.name='Glu2'
Glu2.lower_bound=-1000
Glu2.upper_bound=1000

Glu3=Reaction('Glu3')
Glu3.name='Glu3'
Glu3.lower_bound=-1000
Glu3.upper_bound=1000

transaldolase6=Reaction('transaldolase6')
transaldolase6.name='transaldolase6'
transaldolase6.lower_bound=-1000
transaldolase6.upper_bound=1000
```

```python
####### Metabolite  ###########################


Glucose_6_p = Metabolite(
    'Glucose_6_p',
    formula='C6H12O6',
    name='Glucose_6_p',
    compartment='c')

p6_phosphogluconolactone=Metabolite(
    'p6_phosphogluconolactone',
    formula='C6H13O9P',
    name='p6_phosphogluconolactone',
    compartment='c')

p6_phosphogluconate=Metabolite(
    'p6_phosphogluconate',
    formula='C10H12N5O13P3',
    name='p6_phosphogluconate',
    compartment='c')

Ribulose_5_phosphate=Metabolite(
    'Ribulose_5_phosphate',
    formula='C10H15N5O10P2',
    name='Ribulose_5_phosphate',
    compartment='c')

ribulose_5_phosphate=Metabolite(
    'ribulose_5_phosphate',
    formula='C10H15N5O10P2',
    name='ribulose_5_phosphate',
    compartment='c')


xylulose_5_phospate=Metabolite(
    'xylulose_5_phospate',
    formula=' H',
    name='xylulose_5_phospate',
    compartment='c')


Glyceraldehyde_3_phosphate= Metabolite(
    'Glyceraldehyde_3_phosphate + ',
    formula='C6H13O9P',
    name='Glyceraldehyde_3_phosphate',
    compartment='c')

sedoheptulose_7_phosphate= Metabolite(
    'sedoheptulose_7_phosphate',
    formula='C6H13O9P',
    name='sedoheptulose_7_phosphate',
    compartment='c')


Fructose_6_phosphate = Metabolite(
    'Fructose_6_phosphate + ',
    formula='C6H14O12P2',
    name='Fructose_6_phosphate',
    compartment='c')

Erythrose_4_phosphate = Metabolite(
    'Erythrose_4_phosphate',
    formula='C3H7O6P',
    name='Erythrose_4_phosphate',
    compartment='c')

xylulose_5_phosphate =Metabolite(
    'xylulose_5_phosphate',
    formula='C3H7O6P',
    name='xylulose_5_phosphate',
    compartment='c')

Glyceraldehyde_3_phosphate = Metabolite(
    'Glyceraldehyde_3_phosphate',
    formula='C3H8O10P2',
    name='Glyceraldehyde_3_phosphate +',
    compartment='c')

Fructose_6_phosphate = Metabolite(
    'Fructose_6_phosphate',
    formula='C3H7O7P',
    name='Fructose_6_phosphate',
    compartment='c')

Glycolsis = Metabolite(
    'Glycolsis',
    formula='C3H7O7P',
    name='Glycolsis',
    compartment='c')
```

```python
############# add Metabolites To Reactions ##################################3


Glucose_6_phosphate_dehydrogenase.add_metabolites({
     Glucose_6_p:-1.0,
     p6_phosphogluconolactone:1.0
})


Gluconolactonase.add_metabolites({
     p6_phosphogluconolactone:-1.0,
     p6_phosphogluconate:1.0
})


p6_phosphogluconate_dehydratenase.add_metabolites({
    p6_phosphogluconate:-1,
    Ribulose_5_phosphate :1 
})

ribulose_5_phosphate_3_Epimerase.add_metabolites({
    Ribulose_5_phosphate:-1,
    xylulose_5_phosphate:1
})


ribulose_5_phosphate_isomerase.add_metabolites({
    Ribulose_5_phosphate:-1,
    ribulose_5_phosphate :1
})

transletolase1.add_metabolites({
    ribulose_5_phosphate:-1 ,
    Glyceraldehyde_3_phosphate:1
    
})

transletolase2.add_metabolites({
    xylulose_5_phosphate:-1,
    sedoheptulose_7_phosphate:1 
})

transaldolase.add_metabolites({
    sedoheptulose_7_phosphate:-1,
    Erythrose_4_phosphate:1
    
})

transaldolase3.add_metabolites({
    Glyceraldehyde_3_phosphate:-1,
    Fructose_6_phosphate:1
})

transaldolase6.add_metabolites({
    Erythrose_4_phosphate:-1,
    Glyceraldehyde_3_phosphate:1
})


transaldolase4.add_metabolites({
    xylulose_5_phosphate:-1,
    Fructose_6_phosphate:1
})

Glu1.add_metabolites({
    Glyceraldehyde_3_phosphate:-1,
    Glycolsis:1 
})

Glu2.add_metabolites({
    Fructose_6_phosphate:-1 
})


Glu3.add_metabolites({
    Fructose_6_phosphate:-1 ,
    Glycolsis:1
})

```

```python
############ Add  Reactions to Model ##########################

model.add_reactions([Glucose_6_phosphate_dehydrogenase,Gluconolactonase,p6_phosphogluconate_dehydratenase,ribulose_5_phosphate_3_Epimerase,ribulose_5_phosphate_isomerase,transletolase1,transletolase2,transaldolase,transaldolase3,transaldolase4,Glu1,Glu2,Glu3,transaldolase6])
```

```python
### Add  objective Function #####################

model.objective = 'Glu2'
```

```python
model.optimize()
```

```python
model.optimize()[:6]
```

```python
### S-Marix #############

import cobra
s=cobra.util.array.create_stoichiometric_matrix(model)
```
